# Scraping des pages linkedin

```python
def Recherche(word,nb_profils):
    
    #CONNECTION
    
    driver = webdriver.Firefox()
    time.sleep(int(rd.randint(1,2)))  
    driver.get('https://www.linkedin.com')
    time.sleep(int(rd.randint(1,2)))  
    
    # locate email form by_class_name
    username = driver.find_element(By.ID, "session_key")
    time.sleep(int(rd.randint(1,2)))  
    # send_keys() to simulate key strokes
    username.send_keys('projet.scraping.esilv@gmail.com')
    time.sleep(int(rd.randint(1,2)))  
    
    # locate password form by_class_name
    password = driver.find_element(By.ID, "session_password")
    time.sleep(int(rd.randint(1,2)))  
    # send_keys() to simulate key strokes
    password.send_keys('MdpScraping')
    time.sleep(int(rd.randint(1,2)))  
    
    # locate submit button by_class_id
    log_in_button = driver.find_element(By.CLASS_NAME, 'sign-in-form__submit-button')
    time.sleep(int(rd.randint(1,2)))  
    # .click() to mimic button click
    log_in_button.click()
    time.sleep(int(rd.randint(1,2)))  
    
    #RECHERCHE
    
    page=0
    nb=0
    url = 'https://www.linkedin.com/search/results/people/?keywords=' + word + '&page=' + str(page)

    L=[]
    while nb<nb_profils:
        search_results = []
        
        page+=1
        url = 'https://www.linkedin.com/search/results/people/?keywords=' + word + '&page=' + str(page)
        
        time.sleep(int(rd.randint(1,2)))   
        driver.get(url)

        time.sleep(int(rd.randint(1,2)))   
        soup=BeautifulSoup(driver.page_source,'html.parser') #html data of the page
        

        search = driver.find_elements(By.CLASS_NAME, "app-aware-link")
  
        for k in search[7:]:
            if k.text != '' and k.text != 'LinkedIn Member':
                search_results.append(k.get_attribute("href"))

        nb+=len(search_results)
        L.extend(search_results)
    
    driver.close() 

    return L
```

# Scraping profils

```python
def Scraping(url):
    driver = webdriver.Firefox()

    time.sleep(int(rd.randint(1,2)))   
    driver.get(url)

    time.sleep(int(rd.randint(1,2)))   
    soup=BeautifulSoup(driver.page_source,'html.parser') #html data of the page
    
    driver.close()  
    return soup
```

```python
def Donnees_profil(soup,url):
    #Infos générales
    prenom=soup.h1.text.strip().split(' ')[0]
    nom=soup.h1.text.strip().split(' ')[1]
    lieu=soup.h3
    if lieu != None:
        lieu = lieu.text.strip().split('\n')[0]
    
    #Expériences
    L_exp=[]
    L_exp_html=soup.find_all('ul', {'class': 'experience__list'})
    if len(L_exp_html)!=0:
        L_exp_html = L_exp_html[0].find_all('li')
        
        if len(L_exp_html)!=0:
            for exp in L_exp_html:
                poste=exp.h3.text.strip()
                entreprise=exp.h4.text.strip()
                duree=exp.find('span', {'class':'date-range__duration'}).text
                L_exp.append([poste,entreprise,duree])
    
    #Formations
    L_form=[]
    L_form_html=soup.find_all('ul', {'class': 'education__list'})
    if len(L_form_html)!=0:
        L_form_html = L_form_html[0].find_all('li')
        
        if len(L_form_html)!=0:
            for form in L_form_html:
                ecole=form.h3.text.strip()
                formation=form.h4.text.strip()
                A1=form.find_all('span', {'class':'date-range'})
                if len(A1)!=0:
                    A1=A1[0].text.strip().split(' - ')[0]
                A2=form.find_all('span', {'class':'date-range'})
                if len(A2)!=0:
                    A2=A2[0].text.strip().split(' - ')[1]
                L_form.append([ecole,formation,A1,A2])
    
    return [prenom,nom,lieu,L_exp,L_form,url]
```

```python
L_profils=[]
for url in L_liens:
    soup=Scraping(url)
    time.sleep(int(rd.randint(1,3)))   
    L_profils.append(Donnees_profil(soup,url))
    
L_profils
```

# Classement profils

```python
def duree_formation(duree):
    if 'an' in duree:
        duree=duree.split(' ')
        annee=int(duree[0])
        
        if len(duree)>2:
            mois=int(duree[2])
            
            return annee*12 + mois
            
        return annee*12        
    else:
        return int(duree.split(' ')[0])       
```

```python
def Classement(ville,deg_ville,formation_voulue,deg_formation,experience_voulue,deg_experience,L_profils):
    #copie de la liste
    L_profils_score=L_profils.copy()

    #Chaque candidat
    for profil in L_profils_score:
        score=0
        
        #Lieu
        if profil[2]!=None:
            if re.search(ville, profil[2],re.I) != None:
                score += deg_ville
        
        #Formations du candidat4
        if len(profil[4])!=0:
            pos_ecole_inge=0
            pos_ecole_comm=0
            for formation in profil[4]:
                df_inge=pd.read_csv('Classement_ingenieurs.csv',sep=';')
                df_comm=pd.read_csv('Classement_commerces.csv',sep=';')

                #on split pour comparer les valeurs aux fichiers csv
                noms_ecole = formation[0].split(' - ')
                for nom in noms_ecole:
                    x=df_inge[df_inge['Ecole'].str.contains(nom, regex= True, na=False, case=False)]
                    #Si une école d'ingénieur correspond, on récupère son classement
                    if len(x)!=0:
                        x=x.reset_index()
                        pos_ecole_inge=x.loc[0,'Rank']

                        if pos_ecole_inge<=10 and formation_voulue=="École d'ingénieur TOP 10":
                            score+=deg_formation
                        elif pos_ecole_inge<=50 and formation_voulue=="École d'ingénieur TOP 50":
                            score+=deg_formation
                        elif pos_ecole_inge<=100 and formation_voulue=="École d'ingénieur TOP 100":
                            score+=deg_formation
                        elif formation_voulue=="École d'ingénieur":
                            score+=deg_formation

                    y=df_comm[df_comm['Ecole'].str.contains(nom, regex= True, na=False, case=False)]
                    #Si une école de commerce correspond, on récupère son classement
                    if len(y)!=0:
                        y=y.reset_index()
                        pos_ecole_comm=y.loc[0,'Rank']

                        if pos_ecole_comm<=5 and formation_voulue=="École de commerce TOP 5":
                            score+=deg_formation
                        elif pos_ecole_comm<=20 and formation_voulue=="École de commerce TOP 20":
                            score+=deg_formation
                        elif formation_voulue=="École de commerce":
                            score+=deg_formation
                        
        #Expériences du candidat
        if len(profil[3])!=0:
            tot_experience=0
            for experience in profil[3]:
                if len(experience)>2:
                    duree_experience=duree_formation(experience[2])
                    tot_experience+=duree_experience

            score+=((tot_experience/12) / experience_voulue) * deg_experience
        
        #print(profil)
        profil.append(score)
      
    L_profils_score.sort(key = lambda x: x[-1], reverse=True)

    return L_profils_score
```

```python
L_classement=Classement(ville,ville_imp,formation,formation_imp,experience,experience_imp,L_profils)
L_classement
```

```python
df_final = pd.DataFrame(L_classement,columns=['Prenom','Nom','Lieu','Experiences','Formations','Lien_linkedin','Score'])
df_final=df_final.drop(['Experiences','Formations'],axis=1)
df_final=df_final.drop_duplicates()
df_final
```
